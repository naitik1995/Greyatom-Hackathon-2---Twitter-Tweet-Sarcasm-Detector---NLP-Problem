```python
lv_string = 'I m so good at focusing while doing homework'
from nltk import word_tokenize
tokens = word_tokenize(lv_string)
```

```python
print(tokens)
```

```python
pos_score = []
neg_score = []
tagged = nltk.pos_tag(tokens)
pos=0
neg=0
obj=0
count=0
count_pos = 0
count_neg = 0
for word, tag in tagged:
    ss_set = None
    ss_set = list(swn.senti_synsets(word))
    if len(ss_set) != 0:
        pos=pos+ss_set[0].pos_score()
        neg=neg+ss_set[0].neg_score()
        obj=obj+ss_set[0].obj_score()
        if pos > neg:
            count_pos += 1
        if neg > pos:
            count_neg += 1
    pos = 0
    neg = 0
    obj = 0
pos_score.append(count_pos)
neg_score.append(count_neg)
```

```python
senti_score = []
pos_score = []
neg_score = []
for i in df_new.Tokenized:
    tagged = nltk.pos_tag(i)
    pos=0
    neg=0
    obj=0
    count=0
    count_pos = 0
    count_neg = 0
    for word, tag in tagged:
        ss_set = None
        ss_set = list(swn.senti_synsets(word))
        if len(ss_set) != 0:
            pos=ss_set[0].pos_score()
            neg=ss_set[0].neg_score()
            if np.abs(pos) > np.abs(neg):
                count_pos += 1
            if np.abs(neg) > np.abs(pos):
                count_neg += 1
        pos = 0
        neg = 0
        obj = 0
    pos_score.append(count_pos)
    neg_score.append(count_neg) 
```
